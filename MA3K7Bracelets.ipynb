```python
import random

n = 100

def generate_bracelet():
    bracelet = []
    num1 = random.randint(0, n)
    num2 = random.randint(0, n)
    bracelet.extend([num1, num2])
    initial_pair = (num1, num2)
    
    while True:
        next_num = (num1 + num2) % (n + 1)
        bracelet.append(next_num)
        if (num2, next_num) == initial_pair:
            break
        num1 = num2
        num2 = next_num
    
    return bracelet[:-2]

def generate_bracelet2(new_pair):
    bracelet2 = []
    num1 = new_pair[0]
    num2 = new_pair[1]
    bracelet2.extend([num1, num2])
    initial_pair = (num1, num2)
    while True: 
        next_num = (num1 + num2) % (n + 1)
        bracelet2.append(next_num)
        if (num2, next_num) == initial_pair:
            break
        num1 = num2
        num2 = next_num
    return bracelet2

def main():
    all_possible_pairs = {(i, j) for i in range(n + 1) for j in range(n + 1)}
    used_pairs = set()
    total_bracelets = 0
    
    while all_possible_pairs:
        remaining_pairs_set = all_possible_pairs - used_pairs
        remaining_pairs = list(remaining_pairs_set)
        if not remaining_pairs:
            break
        
        new_pair = remaining_pairs[0]
        bracelet2 = generate_bracelet2(new_pair)
        
        print(f"Bracelet {total_bracelets + 1}: {','.join(map(str, bracelet2[:-2]))}")
        print(f"Length of Bracelet {total_bracelets + 1}: {len(bracelet2[:-2])}")
        
        for i in range(len(bracelet2)):
            pair = (bracelet2[i], bracelet2[(i + 1) % len(bracelet2)])
            used_pairs.add(pair)
        
        all_possible_pairs.remove(new_pair)
        total_bracelets += 1
    
    print(f"Total number of bracelets generated: {total_bracelets}")
    print("All pairs have been used!")

if __name__ == "__main__":
    main()
```
