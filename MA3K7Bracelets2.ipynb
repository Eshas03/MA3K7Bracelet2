```python
import random
import matplotlib.pyplot as plt

def generate_bracelet(n):
    bracelet = []
    num1 = random.randint(0, n)
    num2 = random.randint(0, n)
    bracelet.extend([num1, num2])
    initial_pair = (num1, num2)
    
    while True:
        next_num = (num1 + num2) % (n + 1)
        bracelet.append(next_num)
        if (num2, next_num) == initial_pair:
            break
        num1 = num2
        num2 = next_num
    
    return bracelet[:-2]

def generate_bracelet2(new_pair, n):
    bracelet2 = []
    num1 = new_pair[0]
    num2 = new_pair[1]
    bracelet2.extend([num1, num2])
    initial_pair = (num1, num2)
    while True: 
        next_num = (num1 + num2) % (n + 1)
        bracelet2.append(next_num)
        if (num2, next_num) == initial_pair:
            break
        num1 = num2
        num2 = next_num
    return bracelet2

def generate_and_count_bracelets(n):
    all_possible_pairs = {(i, j) for i in range(n + 1) for j in range(n + 1)}
    used_pairs = set()
    total_bracelets = 0
    
    while all_possible_pairs:
        remaining_pairs_set = all_possible_pairs - used_pairs
        remaining_pairs = list(remaining_pairs_set)
        if not remaining_pairs:
            break
        
        new_pair = remaining_pairs[0]
        bracelet2 = generate_bracelet2(new_pair, n)
        
        for i in range(len(bracelet2)):
            pair = (bracelet2[i], bracelet2[(i + 1) % len(bracelet2)])
            used_pairs.add(pair)
        
        all_possible_pairs.remove(new_pair)
        total_bracelets += 1
    
    return total_bracelets

def main():
    n_values = list(range(101))  # Values of n from 0 to 100
    bracelet_counts = []
    
    for n in n_values:
        total_bracelets = generate_and_count_bracelets(n)
        bracelet_counts.append(total_bracelets)
    
    plt.plot(n_values, bracelet_counts, marker='o')
    plt.xlabel('n')
    plt.ylabel('Number of Bracelets')
    plt.title('Number of Bracelets vs. n')
    plt.grid(True)
    plt.show()

def check_bracelets_for_n(n):
    total_bracelets = generate_and_count_bracelets(n)
    print(f"Number of Bracelets for n = {n}: {total_bracelets}")

if __name__ == "__main__":
    main()
    
    # Example: Check the number of bracelets for a specific n
    n_to_check = 100 # Replace with the desired n value
    check_bracelets_for_n(n_to_check)
```
